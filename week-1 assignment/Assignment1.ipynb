```python
import yfinance as yf
import pandas as pd
```

```python
stocks = ['AAPL', 'GOOGL', 'MSFT', 'AMZN']
```

```python
start_date = '2018-01-01'
end_date = '2023-01-01'
```

```python
df = pd.DataFrame()
```

```python
for stock_symbol in stocks:
    stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

    stock_data = stock_data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]

    stock_data.columns = [f'{stock_symbol}_{col}' for col in stock_data.columns]

    df = pd.concat([df, stock_data], axis=1)
```

```python
df.to_csv('stock_data.csv', index=True)

print("Data has been saved to stock_data.csv.")
```
